```python
import cv2
import matplotlib.pyplot as plt


lena = plt.imread('/content/LenaB&W (2).jpg')
plt.imshow(lena)
plt.show()

#Calculate frequency of pixels in range 0-255
hist = cv2.calcHist([lena], [0], None, [256], [0, 256])

plt.title("Lena image grayscale histogrtam analysis")
plt.xlabel("Bins")
plt.ylabel("no of Pixels")
plt.plot(hist)
plt.show()
```

```python
# lena Color image histogram analysis of the Gray Scale and RGB

import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np

image = cv2.imread('/content/Lenacolor.jpg', 1)
image = cv2.resize(image, (256, 256))
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(6, 12))

# Image
plt.subplot(3, 1, 1)  # 3 rows, 1 column, 1st subplot
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Grayscale histogram in the middle
plt.subplot(3, 1, 2)  # 3 rows, 1 column, 2nd subplot
hist_gray = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
plt.plot(hist_gray)
plt.title('lena image grayscale histogram analysis')
plt.xlabel('bins')
plt.ylabel('no. of pixels')

# Color histogram at the bottom
plt.subplot(3, 1, 3)  # 3 rows, 1 column, 3rd subplot
color = ('b', 'g', 'r')
for i, col in enumerate(color):
    histr = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
plt.title('lena image RGB Histogram analysis')
plt.xlabel('bins')
plt.ylabel('no. of pixels')

plt.tight_layout()
plt.show()
```

2.Contrast And Brightness Streching

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('/content/LenaB&W (2).jpg', 1)

brightness = 10
contrast = 2

adjusted_image = cv2.convertScaleAbs(image, alpha=contrast, beta=brightness)

# Display original image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Display adjusted image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB))
plt.title('Adjusted Image')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# on Cameraman
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('/content/cameraman (1).jpg', 1)

brightness = 75
contrast = 1.3

adjusted_image = cv2.convertScaleAbs(image, alpha=contrast, beta=brightness)

# Display original image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Display adjusted image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB))
plt.title('Adjusted Image')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

3.Hitsogram equalization

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np

image = cv2.imread('/content/LenaB&W (2).jpg', 0)

equalized_image = cv2.equalizeHist(image)

print("before")
cv2_imshow(image)
print("after")
cv2_imshow(equalized_image)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(hist_original)
plt.title('Original Image Histogram')

plt.subplot(1, 2, 2)
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
plt.plot(hist_equalized)
plt.title('Equalized Image Histogram')

plt.show()
```
